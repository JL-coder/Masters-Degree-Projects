## Q6 [15 points]: How many people embarked on Titanic from a place other than "Southampton"?

```python
# guideline: must use SQL, along with Python
print(pysqldf("SELECT embark_town FROM titanic WHERE embark_town != 'Southampton';").value_counts().sum())
#From this we know that 245 people embarked on the titanic from someplace other that southampton
```

## Q7 [15 points]: How many first-class passengers survived after the accident?
